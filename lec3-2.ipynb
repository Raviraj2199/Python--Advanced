Week Exercise 1: 

Using the yfinance library to extract the closing stock prices for a company.

Calculate and print following statistics: minimum, maximum, mean, median, and standard deviation

Also generate line chart and histogram of prices. 

```python
import yfinance as yf 

txn = yf.Ticker("TXN")

txn_history_data = txn.history("1mo")
close = txn_history_data["Close"]

print(txn_history_data)

print("Type of data=",type(txn_history_data))
```

```python
txn_history_data[["Close","Open","High"]].plot()
```

```python
min =close.min()
print("Min value=", min)
max =close.max()
print("Max value=",max)
mean =close.mean()
print("Mean value=",mean)
median =close.median()
print("Median value=",median)
Standarddeviation =close.std()
print("Standar ddeviation value=",Standarddeviation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=txn_history_data[["Close","Open","High"]])
plt.show()
```
